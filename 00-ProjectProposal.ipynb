## Background and Prior Work

Technology has been invented in order to make human lives easier, but as this technology keeps evolving, the world starts changing too. With the invention of artificial intelligence (AI), many controversies have come up when it comes to the benefits and disadvantages that it brings to the environment, the economy, education, and our everyday life. Some studies suggest that AI can support environmental sustainability by improving efficiency, energy systems and allow better monitoring of environmental conditions. [<a href="#ref1">1</a>] [<a href="#ref2">2</a>] At the same time, the fast expansion of AI has created new forms of resource demand that raise sustainability concerns. As AI data centers increase across economic sections, understanding both its potential benefits and environmental consequences has become very important. 
    
AI systems rely on very large computer infrastructure that is mainly located in data centers in order to support tasks such as cloud computing, machine learning training, and large data processing. Reviews of data center operations indicate that energy intensive servers and cooling systems make these facilities among the fastest growing sources of industrial electricity consumption worldwide. [<a href="#ref3">3</a>] In the United States, the growth of AI has contributed to a fast expansion of data centers infrastructure, increasing overall demand for electricity and placing tension on power systems. [<a href="#ref4">4</a>] As AI continues to grow, the physical infrastructure needed to support these technologies has become a main factor in looking at sustainability. 
    
Studies demonstrate that data centers contribute to a significant and growing environmental burden in the AI era. Guidi et al, find that U.S. data centers consume a major portion of national electricity and that AI data centers are only increasing this. Large chunks of data centers can put tension on regional power grids, increase peak demand, and sometimes prolong dependency on carbon intensive energy sources. [<a href="#ref5">5</a>] In addition to energy consumption, water usage for cooling is a major concern, especially in dry regions. Although solutions like advanced cooling systems, waste heat reuse, and water positive designs have been proposed, their adoption remains inconsistent across facilities. [<a href="#ref6">6</a>] Together, these studies show that AI data centers generate major environmental impacts related to electricity demand, greenhouse emissions, and water consumption. 
    
In addition to environmental concerns, data centers are often promoted as an economic growth impact in host communities. Research indicates that regulatory environments can increase data center valuation, and attract private investment, encouraging companies to locate large facilities in certain regions. [<a href="#ref7">7</a>] Data centers may also contribute to local tax revenue infrastructure development and employment opportunities, particularly in technology and engineering sections. [<a href="#ref8">8</a>] However, community research suggests that these economic benefits are often going along with local costs, including increased demand for electricity and water, land use pressure, and environmental stress. [<a href="#ref9">9</a>] While prior work studies have examined economic outcomes or environmental impacts separately, only a few directly compare these topics across multiple cities. The lack of research motivates a modified evaluation of whether the economic impacts associated with AI data centers outweigh their environmental costs.

### B. Data Storage
 - [ ] **B.1 Data security**: Do we have a plan to protect and secure data (e.g., encryption at rest and in transit, access controls on internal users and third parties, access logs, and up-to-date software)?
 - [ ] **B.2 Right to be forgotten**: Do we have a mechanism through which an individual can request their personal information be removed?
 - [ ] **B.3 Data retention plan**: Is there a schedule or plan to delete the data after it is no longer needed?

### C. Analysis
 - [ ] **C.1 Missing perspectives**: Have we sought to address blindspots in the analysis through engagement with relevant stakeholders (e.g., checking assumptions and discussing implications with affected communities and subject matter experts)?
 - [ ] **C.2 Dataset bias**: Have we examined the data for possible sources of bias and taken steps to mitigate or address these biases (e.g., stereotype perpetuation, confirmation bias, imbalanced classes, or omitted confounding variables)?
 - [ ] **C.3 Honest representation**: Are our visualizations, summary statistics, and reports designed to honestly represent the underlying data?
 - [ ] **C.4 Privacy in analysis**: Have we ensured that data with PII are not used or displayed unless necessary for the analysis?
 - [ ] **C.5 Auditability**: Is the process of generating the analysis well documented and reproducible if we discover issues in the future?

### D. Modeling
 - [ ] **D.1 Proxy discrimination**: Have we ensured that the model does not rely on variables or proxies for variables that are unfairly discriminatory?
 - [ ] **D.2 Fairness across groups**: Have we tested model results for fairness with respect to different affected groups (e.g., tested for disparate error rates)?
 - [ ] **D.3 Metric selection**: Have we considered the effects of optimizing for our defined metrics and considered additional metrics?
 - [ ] **D.4 Explainability**: Can we explain in understandable terms a decision the model made in cases where a justification is needed?
 - [ ] **D.5 Communicate limitations**: Have we communicated the shortcomings, limitations, and biases of the model to relevant stakeholders in ways that can be generally understood?

### E. Deployment
 - [ ] **E.1 Monitoring and evaluation**: Do we have a clear plan to monitor the model and its impacts after it is deployed (e.g., performance monitoring, regular audit of sample predictions, human review of high-stakes decisions, reviewing downstream impacts of errors or low-confidence decisions, testing for concept drift)?
 - [ ] **E.2 Redress**: Have we discussed with our organization a plan for response if users are harmed by the results (e.g., how does the data science team evaluate these cases and update analysis and models to prevent future harm)?
 - [ ] **E.3 Roll back**: Is there a way to turn off or roll back the model in production if necessary?
 - [ ] **E.4 Unintended use**: Have we taken steps to identify and prevent unintended uses and abuse of the model and do we have a plan to monitor these once the model is deployed?
