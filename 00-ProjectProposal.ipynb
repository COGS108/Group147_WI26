## Research Question


Research Question: AI data centers have expanded rapidly in recent years, increasing corporate productivity, creating new job industries and enabling innovations such as AI chatbots with 24/7 customer service. However, the growth of AI data centers around the country and so has their demand for local resources, specifically energy and water usage. By evaluating the economic benefits associated with AI data centers and the environmental costs they impose in host cities, to what extent do these benefits outweigh the environmental costs across different cities? 



## Background and Prior Work

Technology has been invented in order to make human lives easier, but as this technology keeps evolving, the world starts changing too. With the invention of artificial intelligence (AI), many controversies have come up when it comes to the benefits and disadvantages that it brings to the environment, the economy, education, and our everyday life. Some studies suggest that AI can support environmental sustainability by improving efficiency, energy systems and allow better monitoring of environmental conditions12. At the same time, the fast expansion of AI has created new forms of resource demand that raise sustainability concerns. As AI data centers increase across economic sections, understanding both its potential benefits and environmental consequences has become very important. 
 AI systems rely on very large computer infrastructure that is mainly located in data centers in order to support tasks such as cloud computing, machine learning training, and large data processing. Reviews of data center operations indicate that energy intensive servers and cooling systems make these facilities among the fastest growing sources of industrial electricity consumption worldwide.3 In the United States, the growth of AI has contributed to a fast expansion of data centers infrastructure, increasing overall demand for electricity and placing tension on power systems.4 As AI continues to grow, the physical infrastructure needed to support these technologies has become a main factor in looking at sustainability. 
Studies demonstrate that data centers contribute to a significant and growing environmental burden in the AI era. Guidi et al, find that U.S. data centers consume a major portion of national electricity and that AI data centers are only increasing this. Large chunks of data centers can put tension on regional power grids, increase peak demand, and sometimes prolong dependency on carbon intensive energy sources.5 In addition to energy consumption, water usage for cooling is a major concern, especially in dry regions. Although solutions like advanced cooling systems, waste heat reuse, and water positive designs have been proposed, their adoption remains inconsistent across facilities.6 Together, these studies show that AI data centers generate major environmental impacts related to electricity demand, greenhouse emissions, and water consumption. 
In addition to environmental concerns, data centers are often promoted as an economic growth impact in host communities. Research indicates that regulatory environments can increase data center valuation, and attract private investment, encouraging companies to locate large facilities in certain regions.7 Data centers may also contribute to local tax revenue infrastructure development and employment opportunities, particularly in technology and engineering sections.8 However, community research suggests that these economic benefits are often going along with local costs, including increased demand for electricity and water, land use pressure, and environmental stress.9 While prior work studies have examined economic outcomes or environmental impacts separately, only a few directly compare these topics across multiple cities. The lack of research motivates a modified evaluation of whether the economic impacts associated with AI data centers outweigh their environmental costs.
